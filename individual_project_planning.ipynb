```r
# Load the tidyverse package and read the players data set.
library(tidyverse)
players <- read_csv("data/players.csv")
players
```

```r
# Find the number of observations in the players data set.
count(players)
```

```r
# Produce summary statistics by finding the number of players in each level of gaming experience.
players_experience <- players |>
    group_by(experience) |>
    count()
players_experience
```

```r
# Produce summary statistics by finding the proportion of players subscribed and unsubscribed to the game-related newsletter.
players_subscribe <- players |>
    group_by(subscribe) |>
    count()
players_subscribe
```

```r
# Produce summary statistics by determining if any emails are identical.
players_email <- players |>
    group_by(hashedEmail) |>
    count()
players_email
```

```r
# Calculate summary statistics by finding the average hours played by the players.
players_avg_hours <- players |>
    summarize(played_hours = mean(played_hours, na.rm = TRUE))
players_avg_hours

players_max_hours <- players |>
    summarize(played_hours = max(played_hours, na.rm = TRUE))
players_max_hours

players_min_hours <- players |>
    summarize(played_hours = min(played_hours, na.rm = TRUE))
players_min_hours
```

```r
# Produce summary statistics by determining the occurence of each name.
players_name <- players |>
    group_by(name) |>
    count()
players_name
```

```r
# Produce summary statistics by finding the genders present in the players data set.
players_gender <- players |>
    group_by(gender) |>
    count()
players_gender
```

```r
# Calculate summary statistics by finding the average age of players in the data set.
players_age <- players |>
    summarize(Age = mean(Age, na.rm = TRUE))
players_age

players_max_age <- players |>
    summarize(Age = max(Age, na.rm = TRUE))
players_max_age

players_min_age <- players |>
    summarize(Age = min(Age, na.rm = TRUE))
players_min_age
```

```r
# Read the sessions data set.
sessions <- read_csv("data/sessions.csv")
sessions
```

```r
count(sessions)
```

```r
# Provide summary statistics for the sessions data set by finding the number of session occurrences by email.
sessions_email <- sessions |>
    group_by(hashedEmail) |>
    count()
sessions_email

sessions_email_max <- sessions_email |>
    ungroup() |>
    summarize(n = max(n))
sessions_email_max
```

```r
# Provide summmary statistics by determining if any sessions started or ended at the same time and the number that did.
sessions_start <- sessions |>
    group_by(start_time) |>
    count()
sessions_start

sessions_start_max <- sessions_start |>
    ungroup() |>
    summarize(n = max(n))
sessions_start_max

sessions_end <- sessions |>
    group_by(end_time) |>
    count()
sessions_end

sessions_end_max <- sessions_end |>
    ungroup() |>
    summarize(n = max(n))
sessions_end_max
```

```r
# Calculate the average start and end time for sessions using the original times.
avg_original_start <- sessions |>
    summarize(original_start_time = mean(original_start_time))
avg_original_start

avg_original_end <- sessions |>
    summarize(original_end_time = mean(original_end_time, na.rm = TRUE))
avg_original_end
```

## Exploratory Visualizations

```r
# Create a bar graph comparing the proportion of subscribed players to unsubcribed.
experience_plot <- ggplot(players_subscribe, aes(x = subscribe, y = n)) +
    geom_bar(stat = "identity", fill = "lightpink") +
    labs(x = "Subscription Status", y = "Number of Players", title = "Player Subscription") +
    theme(text = element_text(size = 15)) 
experience_plot
```

```r
# Plot the relationship between age and hours played, colored by subscribe.
age_hours_plot <- ggplot(players, aes(x = Age, y = played_hours, color = subscribe)) +
    geom_point() +
    labs(x = "Age (Years)", y = "Hours Played", title = "Relationship Between Age and Hours Played", color = "Experience Level") +
    theme(text = element_text(size = 13))
age_hours_plot
# It looks like there are only a few points since the range of hours played is so great. 
# If you look closely, you can see clusters of points at common ages of Minecraft players.
```
